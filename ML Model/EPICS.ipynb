```python
import yaml

yaml_path = '/content/local_dataset/data.yaml'

# Load the existing yaml
with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# Point to the new local paths
data['train'] = '/content/local_dataset/train/images'
data['val']   = '/content/local_dataset/valid/images'
data['test']  = '/content/local_dataset/test/images'

# Enforce single class setup
data['nc'] = 1
data['names'] = ['garbage']

# Save it back
with open(yaml_path, 'w') as f:
    yaml.safe_dump(data, f)

print('✅ Successfully updated data.yaml')
```

```python
from ultralytics import YOLO

# YOLOv8s is a strong baseline for Colab's T4 GPU
model = YOLO('yolov8s.pt')

# Train with extensive data augmentations
model.train(
    data='/content/local_dataset/data.yaml', # Point to local yaml
    epochs=100,
    imgsz=640,
    batch=16,
    device=0,
    name='impact_garbage_v2',
    optimizer='AdamW',
    mosaic=1.0,
    mixup=0.15,
    degrees=10.0,
    patience=25
)
```
